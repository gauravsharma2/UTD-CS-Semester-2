```
# /FileStore/tables/alice.txt
alice = sc.textFile("/FileStore/tables/alice.txt")
alice.collect()
```

```
import nltk
from nltk.corpus import stopwords
from nltk.tag import pos_tag
nltk.download('all')
import re
```

```
alice = alice.flatMap(lambda x: re.split("[\[\]# ./\"'-:;|?*“”’‘]", x)).filter(lambda x: len(x)>0)
alice.collect()
```

```
alice_wo_sw = alice.filter(lambda x: x.lower() not in stopwords.words("english"))
alice_wo_sw.collect()
```

```
named_entity = alice_wo_sw.filter(lambda x: nltk.pos_tag([x])[0][1] == "NN")
named_entity.collect()
```

```
pairs = named_entity.map(lambda x: (x.lower(), 1))
pairs.collect()
```

```
count = pairs.reduceByKey(lambda x,y: x+y).sortBy(lambda x: -x[1])
count.collect()
```
